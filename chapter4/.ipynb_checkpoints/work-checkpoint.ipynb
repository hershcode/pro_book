```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import (
    r2_score, 
    mean_squared_error as mse,
    mean_absolute_percentage_error as mape)
import numpy as np
%config InlineBackend.figure_format = 'retina'
```

```python
train = pd.read_excel(io='price_demand_data.xlsx', sheet_name='train')
test = pd.read_excel(io='price_demand_data.xlsx', sheet_name='test')
```

```python
train['log_sales'] = np.log(train['sales'])
train['log_price'] = np.log(train['price'])

test['log_sales'] = np.log(test['sales'])
test['log_price'] = np.log(test['price'])
```

```python
coef_df = pd.DataFrame(index=['linear', 'exponential', 'constant_elasticity'],
                      columns=['D', 'b', 'r_squared'])
coef_df.index.name = 'model'
```

# Linear 

$$ d(p) = D(1-bp)^{+} $$

$ d(p)$ &#8594; demand estimate for product\
$ D $ &#8594; Price response function parameter i.e. constant in linear model \
$ b $ &#8594; Price sensitivity estimate i.e. slope\
$ p $ &#8594; input price

Estimate $D$ and $b$ using linear regression from scikit learn. \
Distribute $D$ to $(1-bp)$, you'll get $ d(p) = D - bDp $. \
You'll have the equation in the form of $y = mx + C$, where $C = D$ and $m = -bD$

After you fit the data using scikit-learn's LinearRegression model the `intercept_` and `coef_` will represent $C$ and $m$, respectively.\
Given our previous equation, our estimate for $D = C$ and $b = \frac{m}{-D}$.

```python
linear_price_model = LinearRegression()
X = train['price'].values.reshape(-1,1)
y = train['sales'].values.reshape(-1,1)
linear_price_model.fit(X,y)
d = round(linear_price_model.intercept_.item(), 2)
b = round(linear_price_model.coef_.item()/(-linear_price_model.intercept_.item()), 2)

test['linear_pred'] = test['price'].apply(lambda p: d*(1-b*p))

coef_df.loc['linear', 'D'] = round(linear_price_model.intercept_.item(), 2)
coef_df.loc['linear', 'b'] = round(linear_price_model.coef_.item()/(-linear_price_model.intercept_.item()), 2)
coef_df.loc['linear','r_squared'] = round(r2_score(y_true=test['sales'], 
                                    y_pred=test['linear_pred']), 2)
```

# Exponential 

$$ d(p) = Ae^{-bp} $$

Estimate A and b using linear regression model from scikit-learn.\
First convert exponential function to linear equation via natural log transformation.\
You'll get, $$ ln[d(p)] = ln[A] - bp $$
We'll estimate $b$ and $ln[A]$ from linear regression. 

```python
linear_exponential_model = LinearRegression()
X = train['price'].values.reshape(-1,1)
y = train['log_sales'].values.reshape(-1,1)
linear_exponential_model.fit(X,y)

d = round(np.exp(linear_exponential_model.intercept_.item()), 2)
b = round(-linear_exponential_model.coef_.item(), 2)

test['exponential_pred'] = test['price'].apply(lambda p: d*np.exp(-b*p))

coef_df.loc['exponential', 'D'] = d
coef_df.loc['exponential', 'b'] = b
coef_df.loc['exponential','r_squared'] = round(r2_score(y_true=test['sales'], 
                                    y_pred=test['exponential_pred']), 2)
```

# Constant elasticity 

$$ d(p) = Ap^{-\epsilon} $$

Apply natural log transformation and apply linear regression to equation.\
We get the following equation: $$ln[d(p)] = ln[A] - \epsilon ln[p]$$

We'll estimate $ln[A]$ and $\epsilon ln[p]$ using scikit-learn LinearRegression

```python
constant_elascticity_model = LinearRegression()
X = train['log_price'].values.reshape(-1,1)
y = train['log_sales'].values.reshape(-1,1)
constant_elascticity_model.fit(X,y)
A = round(np.exp(constant_elascticity_model.intercept_.item()), 2)
epsilon = round(-constant_elascticity_model.coef_.item(), 2)

test['elasticity_pred'] = test['price'].apply(lambda p: A*p**-epsilon)

coef_df.loc['constant_elasticity', 'D'] = A
coef_df.loc['constant_elasticity', 'b'] = epsilon
coef_df.loc['constant_elasticity','r_squared'] = round(r2_score(y_true=test['sales'], 
                                    y_pred=test['elasticity_pred']), 2)
```

# Calculate RMSE 

```python
coef_df.loc['linear', 'rmse'] = mse(y_true=test['sales'],
                                   y_pred=test['linear_pred'],
                                   squared=False)

coef_df.loc['exponential', 'rmse'] = mse(y_true=test['sales'],
                                   y_pred=test['exponential_pred'],
                                   squared=False)

coef_df.loc['constant_elasticity', 'rmse'] = mse(y_true=test['sales'],
                                   y_pred=test['elasticity_pred'],
                                   squared=False)
```

# Calculate MAPE

```python
coef_df.loc['linear', 'mape'] = mape(y_true=test['sales'],
                                   y_pred=test['linear_pred'])

coef_df.loc['exponential', 'mape'] = mape(y_true=test['sales'],
                                   y_pred=test['exponential_pred'])

coef_df.loc['constant_elasticity', 'mape'] = mape(y_true=test['sales'],
                                   y_pred=test['elasticity_pred'])
```

# Weighted MAPE 

```python
coef_df.loc['linear', 'weighted_mape'] = mape(y_true=test['sales'],
                                               y_pred=test['linear_pred'],
                                              sample_weight=test['sales'])

coef_df.loc['exponential', 'weighted_mape'] = mape(y_true=test['sales'],
                                                   y_pred=test['exponential_pred'],
                                                  sample_weight=test['sales'])

coef_df.loc['constant_elasticity', 'weighted_mape'] = mape(y_true=test['sales'],
                                                           y_pred=test['elasticity_pred'],
                                                          sample_weight=test['sales'])
```

# Plotting price response curve

```python
from functools import partial

def linear(D, b, p):
    return D*(1-b*p)

def exponential(A, b, p):
    return np.exp(A - b*p)

def elasticity(A, epsilon, p):
    return np.exp(A - epsilon*np.log(p))

def mk_linear(D, b):
    return partial(linear, D, b)

def mk_exponential(A, b):
    return partial(exponential, A, b)

def mk_elasticity(A, epsilon):
    return partial(elasticity, A, epsilon)


l = mk_linear(D=222.02, b=.13)
e = mk_exponential(A=15.65, b=1.88)
ce = mk_elasticity(A=27.4, epsilon=12.82)

import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Sample data
samples = pd.DataFrame(data={'price': np.arange(6, 7.51, 0.01)})
samples['linear'] = samples['price'].apply(lambda x: l(x))
samples['exponential'] = samples['price'].apply(lambda x: e(x))
samples['elasticity'] = samples['price'].apply(lambda x: ce(x))

# Create the plot
ax = samples.set_index('price').plot(xlim=[6, 7.51])

# Add regional shading from 6.45 to 7.35
ax.fill_betweenx(y=[ax.get_ylim()[0], ax.get_ylim()[1]], x1=6.45, x2=7.35, color='gray', alpha=0.3)

# Add a label to the shaded region
ax.text(6.45 + (7.35-6.45)/2.5, ax.get_ylim()[1] - 10, 'Test', fontsize=12, color='black')

# Set line styles for exponential and elasticity curves
ax.lines[1].set_linestyle('--')  # Set the exponential curve to dashed
ax.lines[2].set_linestyle(':')   # Set the elasticity curve to jagged

# Customize labels, title, and legend if needed
plt.ylabel('Predicted Demand')
plt.title('Estimated Price Response')

# Show the legend
plt.legend(['Linear', 'Exponential', 'Elasticity'])

def format_dollars(x, pos):
    dollars = int(x)
    cents = int((x - dollars) * 100)
    return f"${dollars}.{cents:02d}"

ax.xaxis.set_major_formatter(FuncFormatter(format_dollars))
plt.xlabel('Price')

# Set x-axis ticks at 25-cent intervals
ax.set_xticks(np.arange(6, 7.51, 0.25))

# Display the plot
plt.show()
```

# Calculate optimal price

To estimate optimal price we have to maximize a profit function. If we assume the cost of each unit is \$6, then our profit function is:

$$ P = (p - c) * d(p) $$ 

where $P$ is profit, $p$ is price, $c$ is cost which is 6, and $d(p)$ is estimated demand from our price response function. To get the optimal price, you'll find the index where the maximum profit occurs for each price response model. 

```python
# Calculate the profit for each demand function and price
samples['profit_linear'] = (samples['price'] - 6) * samples['linear']
samples['profit_exponential'] = (samples['price'] - 6) * samples['exponential']
samples['profit_elasticity'] = (samples['price'] - 6) * samples['elasticity']

# Determine the optimal price for each demand function
coef_df.loc['linear', 'optimal_price'] = samples['price'][samples['profit_linear'].idxmax()]
coef_df.loc['exponential', 'optimal_price'] = samples['price'][samples['profit_exponential'].idxmax()]
coef_df.loc['constant_elasticity', 'optimal_price'] = samples['price'][samples['profit_elasticity'].idxmax()]

coef_df
```

# Incorporating Potential Demand

Potential demand is the number of customer's who would be interested in the product you're selling. This potential demand represents customers of all types who are willing to buy at different price points. Typically for brick and mortars it is hard to measure potential demand because they only have data on the sales that happened and no data on customer's who considered buying the product and did not purchase. Meanwhile, online retailers know exactly who has look at a product and who has converted. Online retailers can use this viewing data to measure potential demand. This can also be called unconstrained demand.

When we have potential demand available, we can decompose the price-respone function into potential demand (always assumed to be independent of price) and a conversation rate (which incorporates customer response to price).

$$ d_{ijkt} = D_{ijkt} \times \rho_{ijk}(p_{t},\textbf{x}_{t};\boldsymbol{\theta}) + \epsilon_{ijkt} $$

$i$ &#8594; product\
$j$ &#8594; customer segment\
$k$ &#8594; channel\
$t$ &#8594; time period

$D$ &#8594; potential demand\
$\rho$ &#8594; price response function that estimates the fraction of potential demand that converts into sales as a function of price and other variables\
$x$ &#8594; indpendent variables that influence demand\
$\theta$ &#8594; parameters of price response model\
$\epsilon$ &#8594; error term

For the case of a single product and time period, we can drop the subscripts and write the price-response function as: $$ d(p) = D\rho(p) + \epsilon$$

When we know potential demand, we then just have to estimate the parameters of $\rho(p)$ which is the conversion rate function based on empirical observations of the conversion $$\hat{r}_{t} = \frac{\hat{d}_{t}}{\hat{D}_{t}}$$

and the prices $p_{t}$ for values in the training set. 

We use the logit function and brand it the logit converstion-rate function. $$\rho(p) = \frac{e^{a-bp}}{1+e^{a-bp}}$$

We want to find the parameters $a$ and $b$ in sthis function that best fits the conversion rates found from the training set and the generalize well on the test. 

With some algebra we do the following steps:

$$ e^{a-bp} = \frac{\rho(p)}{1 - \rho(p)}$$
$$ a-bp = ln[\frac{\rho(p)}{1 - \rho(p)}]$$
we can then replace with $\rho(p)$ with $\hat{r}_{t}$ and we get, $$ ln(\frac{\hat{r}_{t}}{1-\hat{r}_{t}}) = a - bp_{t}$$

We can apply linear regression to this and get estimates for $a$ and $b$. 

```python
# Create conversation rate column in train and test set 

train['conversion_rate'] = train['sales']/train['potential_demand']
test['conversion_rate'] = test['sales']/test['potential_demand']

train['log_conversion_rate_odds'] = np.log(train['conversion_rate']/(1-train['conversion_rate']))
test['log_conversion_rate_odds'] = np.log(test['conversion_rate']/(1-test['conversion_rate']))

model = LinearRegression()
X = train['price'].values.reshape(-1,1)
y = train['log_conversion_rate_odds'].values.reshape(-1,1)

model.fit(X,y)

logit_params = {
    'a':round(model.intercept_.item(),2),
    'b':round(-model.coef_.item(),2)
}


def logit_model(a, b, p):
    numerator = np.exp(a-b*p)
    denominator = 1 + np.exp(a-b*p)
    return numerator/denominator

def mk_logit_model(a, b):
    return partial(logit_model, a, b)

def price_response(a, b, p, D):
    
    return D * logit_model(a, b, p)

def mk_price_response(a, b):
    return partial(price_response, a, b)

pr = mk_price_response(a=logit_params.get('a'),
                       b=logit_params.get('b'))

test['logit_pred'] = test[['price', 'potential_demand']].apply(lambda row: pr(row['price'],row['potential_demand']), axis=1)

logit_rmse = mse(y_true=test['sales'], y_pred=test['logit_pred'], squared=False)
logit_mape = mape(y_true=test['sales'], y_pred=test['logit_pred'])
logit_weighted_mape = mape(y_true=test['sales'], y_pred=test['logit_pred'], sample_weight=test['sales'])

print(f"Test RMSE: {logit_rmse}")
print(f"Test MAPE: {logit_mape}")
print(f"Test Weighted MAPE: {logit_weighted_mape}")
```

In summary, we applied linear regression to data to estimate parameters of the logit model. That estimated logit model would give us the conversion rate of prices and we then multiply that by potential demand to get estimate of number of sales. 

We can get the optimal price by doing running our profit calculation but we'd need a potential demand estimate and then we could get an appropriate demand estimate which we could plug into our profit objective function.

$$ d(p) = D(1-bp-cw) $$

Another way to model this via an additive interaction with elasticity. The form of this model would be: $$ d(p) = Dp^{-b + cw}$$
This additive interaction with elasticity implies that the

```python
# Constant-Elasticity 
train['log_weekend_plus_one'] = np.log(train['weekend'] + 1)
multi_elasticity_model = LinearRegression()
X = train[['log_price', 'log_weekend_plus_one']].values
y = train['log_sales'].values.reshape(-1,1)
multi_elasticity_model.fit(X,y)

D = round(np.exp(multi_elasticity_model.intercept_.item()), 2)
b = round(-multi_elasticity_model.coef_[0][0], 2)
c = round(-multi_elasticity_model.coef_[0][1], 2)

test['multi_elasticity_pred'] = test[['price', 'weekend']].apply(lambda x: D*x['price']**(-b)*(x['weekend'] + 1)**(-c), axis=1)

multi_coef_df.loc['constant_elasticity', 'D'] = d
multi_coef_df.loc['constant_elasticity', 'b'] = b
multi_coef_df.loc['constant_elasticity', 'c'] = c
multie_exponential_r2 = r2_score(y_true=test['sales'], y_pred=test['multi_exponential_pred'])

multi_coef_df.loc['constant_elasticity','r_squared'] = round(multie_exponential_r2, 2)
```

we can then replace with $\rho(p,w)$ with $\hat{r}_{t}$ and we get, $$ ln(\frac{\hat{r}_{t}}{1-\hat{r}_{t}}) = a - bp_{t}-cw$$

We can apply linear regression to this and get estimates for $a$ and $b$. 

```python
X = train[['price', 'weekend']].values
y = train['log_conversion_rate_odds'].values.reshape(-1,1)

model = LinearRegression()
model.fit(X,y)

logit_params = {
    'a':round(model.intercept_.item(),2),
    'b':round(-model.coef_[0][0], 2),
    'c':round(-model.coef_[0][1], 2)
}


def logit_model(a, b, c, p, w):
    numerator = np.exp(a-b*p-c*w)
    denominator = 1 + np.exp(a-b*p-c*w)
    return numerator/denominator

def mk_logit_model(a, b, c):
    return partial(logit_model, a, b)

def price_response(a, b, c, p, w, D):
    
    return D * logit_model(a, b, c, p, w)

def mk_price_response(a, b, c):
    return partial(price_response, a, b, c)

pr = mk_price_response(a=logit_params.get('a'),
                       b=logit_params.get('b'),
                       c=logit_params.get('c'))

test['multi_logit_pred'] = test[['price', 'weekend', 'potential_demand']].apply(lambda row: pr(row['price'],row['weekend'],row['potential_demand']), axis=1)

logit_rmse = mse(y_true=test['sales'], y_pred=test['multi_logit_pred'], squared=False)
logit_mape = mape(y_true=test['sales'], y_pred=test['multi_logit_pred'])
logit_weighted_mape = mape(y_true=test['sales'], y_pred=test['multi_logit_pred'], sample_weight=test['sales'])

print(f"Coefficeints for a, b, and c: {logit_params.get('a')}, {logit_params.get('b')}, {logit_params.get('c')}")

print(f"Test RMSE: {logit_rmse}")
print(f"Test MAPE: {logit_mape}")
print(f"Test Weighted MAPE: {logit_weighted_mape}")
```

From the results, the test RMSE and MAPE improve by 2% while Weighted MAPE is about the same. The coefficients b and c are interesting because they have opposite signs. If you plug in the coefficients to the model it shows:

$$\rho(p,w) = \frac{e^{36.4-5.22p+0.16w}}{1+e^{36.4-5.22p+0.16w}}$$

The -ve coefficient for $p$ indicates for every unit increase in price, the log odds of conversion will decrease. The +ve coefficient for $w$ indicates when it is the weekend, the log odds of conversion rate will increase.
